```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,5)

print("Libraries loaded successfully.")
print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version : {np.__version__}")
```

```python
#load real dataset kaggle.com/datasets/mlg-ulb/creditcardfraud
df = pd.read_csv('creditcard.csv')
print(f"Dataset loaded : {df.shape[0]:,} rows, {df.shape[1]} columns")
print(f"Fraud transactions : {df['Class'].sum():,} ({df['Class'].mean()*100:.2f}%)")
print(f"\nColumns names :")
print(df.columns.tolist())
print(f"\nFirst 5 rows : ")
df.head(5)
```

```python
#Data Quality Check

print("--- Missing Values---")
print(df.isnull().sum().sum(), "total missing values")

print("\n---Basic Statistics---")
print(df[['Time', 'Amount', 'Class']].describe().round(2))

print("\n--- Class Distribution ---")
print(f"Normal transactions : {(df['Class']==0).sum():,} ({(df['Class']==0).mean()*100:.2f}%)")
print(f"Fraud transactions  : {(df['Class']==1).sum():,} ({(df['Class']==1).mean()*100:.2f}%)")
```

```python
#Visualization 1 : Class Distribution
fig, axes = plt.subplots(1, 2, figsize = (14,5))

#نمودار 1 - تعداد 
axes[0].bar(['Normal', 'Fraud'],
           [df['Class'].value_counts()[0], df['Class'].value_counts()[1]],
            color = ['#2ecc71', '#e74c3c'])
axes[0].set_title('Transaction Count by Class')
axes[0].set_ylabel('Count')
           

# نمودار 2 - توزیع مبلغ
axes[1].hist(df[df['Class']==0]['Amount'], bins=50, alpha=0.6, color='#2ecc71', label='Normal')
axes[1].hist(df[df['Class']==1]['Amount'], bins=50, alpha=0.6, color='#e74c3c', label='Fraud')
axes[1].set_title('Amount Distribution : Normal VS Fraud')
axes[1].set_xlabel('Amount ($)')
axes[1].set_ylabel('Count')
axes[1].legend()


plt.tight_layout()
#plt.savefig('plot1_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

# چون تعداد تراکنش های جعلی کم است در نمودار به حالت عادی نشون نمیده. در سلول بعدی دز مقیاس لگاریتمی ان را نشان میدهیم
```

```python
#Visualization 1 : Class Distribution
fig, axes = plt.subplots(1, 3, figsize = (18,5))

counts = df['Class'].value_counts()
#نمودار 1 - تعداد - (with log scale) 
axes[0].bar(['Normal', 'Fraud'],counts , color = ['#2ecc71', '#e74c3c'])
axes[0].set_yscale('log')  # fraud در مقاس لگاریتمی دیده میشود
axes[0].set_title('Transaction Count (Log Scale)')
axes[0].set_ylabel('Count (log)')
           

# نمودار ۲ — توزیع مبلغ فقط fraud
axes[1].hist(df[df['Class']==1]['Amount'], bins=30, color='#e74c3c', alpha=0.8)
axes[1].set_title('Fraud Amount Distribution')
axes[1].set_xlabel('Amount ($)')
axes[1].set_ylabel('Count')

# نمودار ۳ — مقایسه میانگین مبلغ
avg_amounts = df.groupby('Class')['Amount'].mean()
axes[2].bar(['Normal', 'Fraud'], avg_amounts, color=['#2ecc71', '#e74c3c'])
axes[2].set_title('Average Amount: Normal vs Fraud')
axes[2].set_ylabel('Average Amount ($)')

plt.tight_layout()
plt.savefig('plot1_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print(" Plot saved!")
```

```python
# Visualization 2 : Time Analysis
fig, axes = plt.subplots(1, 2, figsize= (14, 5))

#نمودار 1 - توزیع زمانی normal vs fraud
axes[0].hist(df[df['Class']==0]['Time'], bins=50, alpha= 0.6, color='#2ecc71', label='Normal')
axes[0].hist(df[df['Class']==1]['Time'], bins=50, alpha=0.8, color='#e74c3c', label='Fraud')
axes[0].set_title('Transaction Time Distribution')
axes[0].set_xlabel('Time (seconds)')
axes[0].set_ylabel('Count')
axes[0].legend()

#نمودار 2- در طول زمان fraud
fraud_time = df[df['Class']==1]['Time']
axes[1].scatter(fraud_time,df[df['Class']==1]['Amount'],color='#e74c3c', alpha=0.5, s=20) # نمودار پراکندگی
axes[1].set_title('Fraud : Time VS Amount')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_ylabel('Amount ($)')

plt.tight_layout()
plt.savefig('plt2_time_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("plot saved!")
```

```python
#Visualization 3 : Statistical Analysis

# normal , fraud مقایسه آماری 
normal = df[df['Class']==0]['Amount']
fraud = df[df['Class']==1]['Amount']

#چاپ جدول مقایسه ای با فرمت ستونی 
print("          --- Amount Statistics ---")
print(f"{'Metrics' : <15} {'Normal' : >12} {'Fraud' : >12}") # عنوان ستون هاو چپ و راست چین بودنشون
print("-"*40)
print(f"{'Count' : <15} {len(normal):>12,} {len(fraud):>12,}") # تعداد
print(f"{'Mean' : <15} {normal.mean() :>12.2f} {fraud.mean():>12.2f}") # میانگین
print(f"{'Median' : <15} {normal.median() :>12.2f} {fraud.median():>12.2f}") #میانه
print(f"{'Std' : <15} {normal.std() :>12.2f} {fraud.std():>12.2f}") # انحراف معیار
print(f"{'Max' : <15} {normal.max() :>12.2f} {fraud.max():>12.2f}") 
print(f"{'Min' : <15} {normal.min() :>12.2f} {fraud.min():>12.2f}")

# boxplot رسم - outliers نمایش توزیع و
fig, ax = plt.subplots(figsize=(10, 5))
df.boxplot(column = 'Amount', by = 'Class', ax = ax, boxprops = dict(color='#2ecc71'), medianprops = dict(color='#e74c3c', linewidth = 2))
ax.set_title('Amount Distribution by Class')
ax.set_xlabel('Class (0:Normal , 1:Fraud)')
ax.set_ylabel('Amount ($)')
ax.set_ylim(0, 500) # y محدوده 
plt.suptitle('')
plt.savefig('plot3_statistics.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved!")
```
